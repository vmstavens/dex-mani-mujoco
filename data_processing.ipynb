```python
import math as m
import numpy as np
import math

def angle_to_quaternion(angle_degrees):
    # Convert angle to radians
    angle_radians = math.radians(angle_degrees)

    # Calculate quaternion values
    cos_val = math.cos(angle_radians / 2)
    sin_val = math.sin(angle_radians / 2)

    # Quaternion representation for rotation about the z-axis
    quat = [cos_val, 0, 0, sin_val]

    return quat

# Example usage for -90 degrees
angle_degrees = 180
result_quat = angle_to_quaternion(angle_degrees)
print(result_quat)


```

```python

class B:
    def __init__(self):
        # this should be a ference to the Xa in class A
        self.Xa = 0

    def add(self,x):
        return self.Xa + x

class A:
    def __init__(self):
        self.b = B()
        self.Xa = 0

a = A()
print(a.Xa)
a.b.add(2)
print(a.Xa)
```

```python
class B:
    def __init__(self, a_instance):
        # Reference to the Xa in class A
        self.a_instance = a_instance

    def add(self, x):
        # Modify the Xa attribute in class A
        self.a_instance.Xa += x
        return self.a_instance.Xa

class A:
    def __init__(self):
        self.b = B(self)
        self.Xa = 0

a = A()
print(a.Xa)  # Output: 0
a.b.add(2)
print(a.Xa)  # Output: 2
```
